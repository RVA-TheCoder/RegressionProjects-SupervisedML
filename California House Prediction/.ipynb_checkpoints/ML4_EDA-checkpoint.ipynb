### STEP 3 : Univariate Analysis

```python
# Populate the list of numeric attributes and categorical attributes

from pandas.api.types import is_string_dtype, is_numeric_dtype

num_list = []
cat_list = []

for column in data:
    
    if is_numeric_dtype(data[column]):
        num_list.append(column)
        
    elif is_string_dtype(data[column]):
        cat_list.append(column)
        
print('Numerical attributes : ',num_list,'\n')

print('Categorical attributes : ',cat_list)
```

### (i) Histogram and barchart for numerical and categorical attributes respectively.

```python
# Defining subplot layout

total_size = len(num_list) + len(cat_list)
 
n_cols=2
n_rows= int(np.ceil(total_size/n_cols))

fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols , figsize = (12,18))

i = 0
j = 0
for column in num_list:
    
    data[column].plot(ax = axes[i,j],kind = 'hist' , bins = 30 , title = column , ec = 'black')
    
    if j == n_cols-1:
        i = i + 1
        j = 0
    else:
        j = j + 1

for column in cat_list:
    
    data[column].value_counts().plot(ax = axes[i,j],kind = 'bar' , title = column , ylabel = 'Frequency',ec = 'black')
    
    
    if j == n_cols-1:
        i = i + 1
        j = 0
    else:
        j = j + 1
        
```

### (ii) Boxplot for numerical attributes for finding the outliers.

```python
total_size_num = len(num_list) 
 
n_cols=2
n_rows= int(np.ceil(total_size_num/n_cols))

fig, axes = plt.subplots(nrows=n_rows, ncols=n_cols , figsize = (15,22),gridspec_kw={'height_ratios':[0.5,0.5,0.5,0.5,0.6]})

i = 0
j = 0

for column in num_list:
    
 
    axes[i,j].boxplot(data[column], vert=0) 
    axes[i,j].set_title(column)
    #axes[i,j].set_xlabel('Range')
    axes[i,j].set_ylabel('')
    
    if j == n_cols-1:
        i = i + 1
        j = 0
    else:
        j = j + 1
    
    if i == 2 and j == 0:
        axes[2,0].boxplot(data[column],vert=0 )

# Outlier removal requires different range for each columns.

# for total_rooms column : outlier is value >20000
# for total_bedrooms column : outlier is value >20000
# for population column : outlier is value > 12500

# for households column : outlier is value > 3500
# for median_income column : outlier is value > 13 or 13.5   or 
# we can keep the outliers because it is showing strong correlation with house value.
```

#### From correlation analysis following conclusions have been made:

- so checking them for multicollinarity

- Threshold value of VIF is : VIF >= 10 for dropping a column



```python
data.columns
```

```python
mc_feat = ['total_rooms','total_bedrooms', 'population', 'households']
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant
```

- Syntax : variance_inflation_factor(exog, exog_idx)

Explanation :

One recommendation is that if VIF is greater than 5, then the explanatory variable given by 

'exog_idx' is highly collinear with the other explanatory variables, and 

the parameter estimates will have large standard errors because of this.

```python
df = data.copy()
```

```python
len(df)
```

```python
df.dropna(inplace = True)
```

```python
len(df)
```

```python
# Defining custom function for calculating VIF

def compute_vif(feature_list ,dataframe):
    
    X = dataframe[feature_list]
    
    # Adding constant column in X
    X  = add_constant(X ,prepend=True)
    
    # Creating DF for storing VIF values
    vif = pd.DataFrame()
    vif['variable'] = X.columns
    
    # Dropping any NaN values
    X.dropna(inplace = True)
    
    vif['VIF'] = [variance_inflation_factor(X.values ,i) for i in range(X.shape[1])]
    
    vif = vif[vif['variable'] != 'const']
    
    return vif
```

```python
mc_feat
```

```python
compute_vif(mc_feat ,dataframe = df)
```

```python
# lets remove 'households' column

mc_feat.remove('households')
```

```python
mc_feat
```

```python
compute_vif(feature_list = mc_feat , dataframe  = df)
```

#### Conclusion : we can drop the 'households' column.(if needed)
